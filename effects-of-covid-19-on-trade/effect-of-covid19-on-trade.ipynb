```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import openpyxl #must be installed
from importlib import reload 
plt=reload(plt)
```

### Index

```python
index = np.array(df.index)
index
```

### 2015

```python
# imports
imp2015 = np.array(df.loc[(df.Direction =="Reimpor") & (df.Year == 2015)].Value)
impo2015indx = np.array(df.loc[(df.Direction =="Reimpor") & (df.Year == 2015)].index)
# exports
exp2015 = np.array(df.loc[(df.Direction =="Exports") & (df.Year == 2015)].Value)
expo2015indx = np.array(df.loc[(df.Direction =="Exports") & (df.Year == 2015)].index)
```

### 1. Visualisation of Imports

```python
plt.title("EFFECTS OF COVID19 ON TRADE IMPORTS IN 2015", fontsize=20, fontweight='bold', fontfamily='Arial')
plt.xlabel("INDEX",  fontsize=16, fontweight='bold', fontfamily='Arial' )
plt.ylabel("VALUE (USD)",  fontsize=16, fontweight='bold', fontfamily='Arial')
# 2015 Imports
plt.plot(impo2015indx, imp2015, '-or', markeredgecolor="b")
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["figure.facecolor"] ='#fafafa'
plt.show()
```

### 2. Visualisation of Exports

```python
plt.title("EFFECTS OF COVID19 ON TRADE EXPORTS IN 2015", fontsize=20, fontweight='bold', fontfamily='Arial')
plt.xlabel("INDEX",  fontsize=16, fontweight='bold', fontfamily='Arial' )
plt.ylabel("VALUE (USD)",  fontsize=16, fontweight='bold', fontfamily='Arial')
# 2015 Exports
plt.plot(expo2015indx, exp2015, '-ob', markeredgecolor="r")
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["figure.facecolor"] ='#fafafa'
plt.show()
```

### 3. Visualisation of Imports and Exports

```python
plt.title("EFFECTS OF COVID19 ON TRADE EXPORTS AND IMPORTS IN 2015", fontsize=20, fontweight='bold', fontfamily='Arial')
plt.xlabel("INDEX",  fontsize=16, fontweight='bold', fontfamily='Arial' )
plt.ylabel("VALUE (USD)",  fontsize=16, fontweight='bold', fontfamily='Arial')
# 2015 Exports
plt.plot(expo2015indx, exp2015, '-ob', markeredgecolor="r", label="2015 exports")
# 2015 Imports
plt.plot(impo2015indx, imp2015, '-or', markeredgecolor="b", label="2015 imports")
plt.legend()
plt.rcParams["figure.figsize"] = (10, 5)
plt.rcParams["figure.facecolor"] ='#fafafa'
plt.show()
```
